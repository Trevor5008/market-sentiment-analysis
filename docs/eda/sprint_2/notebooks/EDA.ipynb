```python
ohlcv_df["date"] = pd.to_datetime(ohlcv_df["date"])
idx = ohlcv_df.groupby("ticker")["pmc"].idxmax()
events = ohlcv_df.loc[idx, ["ticker", "date", "pmc"]].rename(
    columns={"ticker": "event_ticker", "date": "event_date", "pmc": "event_pmc"}
).reset_index(drop=True)
```

```python
events
```

```python
ticker_max = ohlcv_df.groupby("ticker")["pmc"].max().rename("ticker_max_pmc")
ticker_max
```

```python
dfn = ohlcv_df.merge(ticker_max, on="ticker", how="left")
dfn["pmc_vs_own_max"] = dfn["pmc"] / dfn["ticker_max_pmc"]
dfn
```

```python
from pandas.tseries.offsets import BDay
#Build ±2 *trading day* window around each event date
offsets = [-2, -1, 0, 1, 2]
event_window = (
    events.assign(key=1)
    .merge(pd.DataFrame({"offset": offsets, "key": 1}), on="key")
    .drop(columns="key")
)
event_window["event_date"] = pd.to_datetime(event_window["event_date"]).dt.normalize()

event_window["date"] = event_window.apply(
    lambda r: r["event_date"] + BDay(int(r["offset"])),
    axis=1
)

#Attach what *all tickers* did on those dates (normalized vs their own max)
dfn["date"] = pd.to_datetime(dfn["date"])
event_window["date"] = pd.to_datetime(event_window["date"])
panel = event_window.merge(
    dfn[["ticker", "date", "pmc", "pmc_vs_own_max"]],
    on="date",
    how="left"
).rename(columns={"ticker": "other_ticker"})

# panel now has:
# event_ticker, event_date, offset, date, other_ticker, pmc, pmc_vs_own_max
```

```python
px.box(dfn, x="ticker", y="pmc_vs_own_max")
```

volatility normally is below 60% of a tickers max, so ticker's elevating to 60%-80% when one ticker reaches it peak appears significant

```python
panel.groupby("offset")["pmc_vs_own_max"].mean()
```

volatility before a peak event is within normal percentages. They elevate after an event.

Is the other tickers spike due to one tickers peak or is this a shared elevation that happens to have a peaked ticker?

```python
comp_df = dfn.pivot(
    index="date",
    columns="ticker",
    values="pmc_vs_own_max"
)
comp_df.columns.name = "Tickers"
comp_df
```

```python
comp_df.mean(axis=1)
```

When I average the means across tickers, vol. seems to elevate and descrease within a day.
