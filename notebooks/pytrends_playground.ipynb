```python
from pytrends.request import TrendReq

#timezone offset in minutes
pytrends = TrendReq(hl = "en-US", tz = 300) #America/New York = 300 minutes
```

```python
kw_list = ["Tesla"] #search term(s)
pytrends.build_payload(
    kw_list,

    #categories filters
    #0 -  no filter | 784 - Business News |1179 - company news | 1240 - company earnings
    cat = 0,

    #today 12-m, now 7-d, 2004-01-01 2020-01-01
    timeframe = "now 1-d",

    #"" - global | "US" | "CA"
    geo = "",

    # Google property | news | images | froogle | youtube
    gprop = ""
)
```

```python
df = pytrends.interest_over_time()
```
